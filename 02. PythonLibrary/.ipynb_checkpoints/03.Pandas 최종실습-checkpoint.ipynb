- 범죄데이터를 활용하여 각 연도별 범죄 증감률을 구해보자



```python
# 1. 라이브러리 불러오기
import pandas as pd
```

```python
# 2. 데이터 불러오기
# 변수명 data2019, data2020, data2021
# 컬럼 중 '관서명' -> 인덱스 설정
# 인코딩 encoding = 'euc-kr'
# 2021년도 '광주지방경찰청'은 drop처리
data2019 = pd.read_csv('data/2019.csv', encoding = 'euc-kr', index_col = '관서명')
data2020 = pd.read_csv('data/2020.csv', encoding = 'euc-kr', index_col = '관서명')
data2021 = pd.read_csv('data/2021.csv', encoding = 'euc-kr', index_col = '관서명')
```

```python
data2021.drop('광주지방경찰청', axis = 0, inplace = True)
```

```python
# 3. 한개의 년도에만 존재하는 경찰청 삭제 (2021년도 -> 광주지방경찰청)
```

```python
data2019
```

```python
# 4. 각 연도별로 범죄 발생수의 합계를 구하여 '총합' 컬럼을 추가하기
```

```python
data1 = data2019.loc[:, '살인' : '폭력']
data2019['총합'] = data1.sum(axis = 1)
data2 = data2020.loc[:, '살인' : '폭력']
data2020['총합'] = data1.sum(axis = 1)
data3 = data2021.loc[:, '살인' : '폭력']
data2021['총합'] = data1.sum(axis = 1)
```

```python
# 5. 각 연도별로 '구분' 컬럼의 값이 '발생건수'인 데이터만 출력하여 변수에 담기
```

```python
count_2019 = data2019[data2019.iloc[:, 0]=='발생건수']
count_2020 = data2020[data2020.iloc[:, 0]=='발생건수']
count_2021 = data2021[data2021.iloc[:, 0]=='발생건수']
count_2019
```

```python
# 6. 각 연도별로 총합 데이터만 추출 - > Series 형태로 출력될 것입니다.
```

```python
sum_2019 = count_2019.iloc[:, 6]
sum_2020 = count_2020.iloc[:, 6]
sum_2021 = count_2021.iloc[:, 6]
sum_2019
```

```python
# 7. 총합데이터 이름 변경해주기 (2019총계, 2020총계, 2021총계)
```
